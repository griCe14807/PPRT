```python
import itertools

def move_sympyplot_to_axis(p, ax):
    backend = p.backend(p)
    backend.ax = ax
    backend.process_series()
    backend.ax.spines["right"].set_color("none")
    backend.ax.spines["bottom"].set_position("zero")
    backend.ax.spines["top"].set_color("none")
    plt.close(backend.fig)
    

# 二つごとに組み合わせの引き算をプロット。
from sympy.plotting import plot_implicit
m, b = symbols("m b")

for combi_group in itertools.combinations(the_df.groupby("ID"), 2):
    
    key_1, group_1 = combi_group[0]
    key_2, group_2 = combi_group[1]
    
    Swirr_1 = 1/ (group_1.T2 * m + b)
    Swirr_2 = 1/ (group_2.T2 * m + b)
    
    f_1 = np.dot(group_1.Por, Swirr_1) - group_1.BVI_Por_cum.iloc[-1]
    f_2 = np.dot(group_2.Por, 1/ (group_2.T2 * m + b)) - group_2.BVI_Por_cum.iloc[-1]
    print(key_1, key_2)
    
    # f(x, y)を図示する
    p_1 = plot_implicit(f_1, (m, 0, 1), (b, 0, 2), show=False)
    p_2 = plot_implicit(f_2, (m, 0, 1), (b, 0, 2), show=False)
    
    fig, ax = plt.subplots()
    move_sympyplot_to_axis(p_1, ax)
    move_sympyplot_to_axis(p_2, ax)
    plt.show()
```
