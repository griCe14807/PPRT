```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os

# my module
current_dir = os.getcwd()
sys.path.append(os.path.join(current_dir, '../'))
import calc_cutoff_BVI
```

```python
the_nmr_df = pd.read_csv(r"D:\tmp\calliance1_nmr.csv")
```

```python
# fig1: cpor vs cperm
fig1 = plt.figure(figsize=(8, 8))
ax1 = fig1.add_subplot(1, 1, 1)
# 
ax1.set_yscale("log")
ax1.set_xlabel("cpor")
ax1.set_ylabel("cperm")

# fig2: cpor vs cutoff value
fig2 = plt.figure(figsize=(8, 8))
ax2 = fig2.add_subplot(1, 1, 1)
ax2.set_xlabel("cpor")
ax2.set_ylabel("cut_coates")

# fig3: cperm vs cutoff value
fig3 = plt.figure(figsize=(8, 8))
ax3 = fig3.add_subplot(1, 1, 1)
ax3.set_xscale("log")
ax3.set_xlabel("cperm")
ax3.set_ylabel("cut_coates")


for key, group in the_nmr_df.groupby("ID"):
    t2_cutoff = calc_cutoff_BVI.calc_t2_cutoff(group)
    cpor = group["cpor"].iloc[0]
    cperm = group["cperm"].iloc[0]
    ID = group["ID"].iloc[0]
    ax1.scatter(cpor, cperm)
    ax1.annotate(ID, (cpor, cperm), size=8)
    
    ax2.scatter(cpor, t2_cutoff)
    ax2.annotate(ID, (cpor, t2_cutoff), size=8)
    
    ax3.scatter(cperm, t2_cutoff)
    ax3.annotate(ID, (cperm, t2_cutoff), size=8)
    

plt.show()
```
